MAP NUMERICAL MONTHS TO MONTH NAMES

```python
# Map numeric months to month names
month_dict = {1:'January', 2:'February', 3:'March', 4:'April', 5:'May', 6:'June', 
              7:'July', 8:'August', 9:'September', 10:'October', 11:'November', 12:'December'}
eskom_fd['month'] = eskom_fd['month'].map(month_dict)
```

THE HOURLY AVERAGE DEMAND THAT NEEDS TO BE SUPPLIED BY ALL RESOURCES THAT CAN BE DISPATCHED BY ESKOM NATIONAL CONTROL.

```python
fig = plt.figure(figsize=(30,10))
# Create a bar plot using Seaborn
sns.lineplot(x='hour', y='Residual Demand',data=eskom_fd)


# Set the plot title and axis labels
plt.title('Hourly Residual Demand')
plt.xlabel('hour')
plt.ylabel('Total Demand MW')
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

ENERGY THAT IS EXPORTED FROM RSA TO NEIGHBORING COUNTRIES VS ENERGY THAT IS IMPORTED INTO RSA FROM NEIGHBORING COUNTRIES.


```python
# Reshape the data into a long format
df_long = pd.melt(eskom_fd, id_vars=['year'], value_vars=['International Exports', 'International Imports'], 
                  var_name='Labels', value_name='total')
fig = plt.figure(figsize=(30,10))

# Create a bar plot using Seaborn
sns.lineplot(x='year', y='total', hue='Labels', data=df_long)

# Set the plot title and axis labels
plt.title('Exported Energy vs Imported Energy')
plt.xlabel('Year')
plt.ylabel('Total Energy (Gwh)')
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

THE DISTRIBUTION OF POWER SOURCES


```python
fig = plt.figure(figsize=(30,10))
# Reshape the data into a long format
df_long = pd.melt(eskom_fd, id_vars=['year'], value_vars=['Wind', 'PV', 'CSP', 'Other RE'], 
                  var_name='Power Sources', value_name='total')



# Calculate the total capacity for each power source
df_pie = df_long.groupby(['Power Sources'])['total'].sum().reset_index()

# Create a pie chart with colors
plt.pie(df_pie['total'], labels=df_pie['Power Sources'], autopct='%1.1f%%')


# Add a title
plt.title('Distribution of Power Sources')

# Show the plot
plt.show()

```

THE AMOUNT AND TYPE OF ENERGY GENERATED BY ESKOM FROM 2018 - 2023

```python
# Reshape the data into a long format
df_long = pd.melt(eskom_fd, id_vars=['year'], value_vars=['Thermal Generation','Dispatchable Generation',
       'Nuclear Generation', 'Eskom Gas Generation', 'Eskom OCGT Generation',
       'Hydro Water Generation', 'Pumped Water Generation'], 
                  var_name='Energy Generated', value_name='Total of Generated Energy')

# Calculate the sum of total generated energy for each energy source
df_sum = df_long.groupby(['year', 'Energy Generated'])['Total of Generated Energy'].sum().reset_index()

# Create a line plot using Seaborn
fig = plt.figure(figsize=(30,10))
sns.lineplot(x='year', y='Total of Generated Energy', hue='Energy Generated', data=df_sum)

# Set the plot title and axis labels
plt.title('Energy Generated from 2018 to 2023')
plt.xlabel('Year')
plt.ylabel('Total of Generated Energy')

# Show the plot
plt.show()
```

EXTRACT TIME FROM TIMESTAMP

```python
import datetime as dt
# Define the format to be used. format1 is the format from to_datetime while format2 is the new output format
format1 ="%m/%d/%Y %I:%M:%S %p"
format2 = "%H:%M:%S"

# Convert and store datetime in new format
eskom_fd['Time']= pd.to_datetime(eskom_fd['Timestamp'], format=format1).dt.strftime(format2)

```
