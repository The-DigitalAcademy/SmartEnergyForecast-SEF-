```python
import requests
import json
import pandas as pd
```

```python
# Initialize an empty DataFrame to store the data
movie_df = pd.DataFrame()

# Iterate through the range of movie IDs
for movie_id in range(1, 800):
    # Make a request for the current movie ID
    url = f"https://api.themoviedb.org/3/movie/{movie_id}?api_key=539ba2f0d8506e0d227fed57bdf4e8b7"
    response = requests.get(url)
    movie_data = response.json()

    # Check if the movie data contains the required keys
    if all(key in movie_data for key in ('title', 'release_date', 'budget', 'revenue', 'runtime', 'vote_average', 'vote_count', 'overview', 'tagline', 'genres', 'production_companies', 'production_countries')):
        # Convert the data to a DataFrame and append it to the overall DataFrame
        movie_row = pd.DataFrame({
            'title': [movie_data['title']],
            'release_date': [movie_data['release_date']],
            'budget': [movie_data['budget']],
            'revenue': [movie_data['revenue']],
            'runtime': [movie_data['runtime']],
            'vote_average': [movie_data['vote_average']],
            'vote_count': [movie_data['vote_count']],
            'overview': [movie_data['overview']],
            'tagline': [movie_data['tagline']],
            'genres': [[genre['name'] for genre in movie_data['genres']]],
            'production_companies': [[company['name'] for company in movie_data['production_companies']]],
            'production_countries': [[country['name'] for country in movie_data['production_countries']]]
        })
        movie_df = movie_df.append(movie_row, ignore_index=True)
    else:
        print(f"Movie data for ID {movie_id} is missing a required key.")

# Display the resulting DataFrame
print(movie_df)
```
