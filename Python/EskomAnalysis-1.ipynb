```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

```

```python
# Define features and target variable
features = eskom.drop('Residual Demand', axis=1)
target = eskom['Residual Demand']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=49)
```

```python
y_train
```

```python
y_test
```

```python
lm = LinearRegression()
```

```python


# Assuming that the datetime column is named 'date'
X_train['Date Time Hour Beginning'] = (X_train['Date Time Hour Beginning'] - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s')

# Now drop the original datetime column
X_train = X_train.drop(columns=['Date Time Hour Beginning'])
```

```python
lm.fit(X_train,y_train)


```

```python
y_pred_train = lm.predict(X_train)
```

```python
from sklearn.metrics import r2_score
mse_train = mean_squared_error(y_train,y_pred_train)
r2_train=r2_score(y_train,y_pred_train)
```

```python
print((mse_train))
print(r2_train)
```
