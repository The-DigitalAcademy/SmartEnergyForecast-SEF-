Do we import more energy or export more energy?


```python
# Reshape the data into a long format
df_long = pd.melt(df, id_vars=['year'], value_vars=['International Exports', 'International Imports'], 
                  var_name='category', value_name='total')
fig = plt.figure(figsize=(30,10))

# Create a bar plot using Seaborn
sns.lineplot(x='year', y='total', hue='category', data=df_long)

# Set the plot title and axis labels
plt.title('Imports and Exports')
plt.xlabel('Year')
plt.ylabel('Inports/Exports (Gwh)')
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
# Reshape the data into a long format
df_long = pd.melt(df, id_vars=['year'], value_vars=['Total UCLF', 'Total PCLF', 'Total OCLF'], 
                  var_name='category', value_name='total')
fig = plt.figure(figsize=(30,10))

# Create a bar plot using Seaborn
sns.barplot(x='year', y='total', hue='category', data=df_long)

# Set the plot title and axis labels

plt.title('Planed and unplaned power cuts BY month')
plt.xlabel('Year')
plt.ylabel('Total Power cuts')
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python

# Reshape the data into a long format
df_long = pd.melt(df, id_vars=['month'], value_vars=['Drakensberg Gen Unit Hours',
       'Palmiet Gen Unit Hours', 'Ingula Gen Unit Hours'], 
                  var_name='Power Station', value_name='total')
colors=['blue','orange','green']
# Calculate the total units generated for each power station
df_pie = df_long.groupby(['Power Station'])['total'].sum().reset_index()
fig = plt.figure(figsize=(30,10))

# Create a pie chart
plt.pie(df_pie['total'], labels=df_pie['Power Station'], autopct='%1.1f%%',colors=colors)

# Add a title
plt.title('Total Units Generated by Power Station')

# Show the plot
plt.show()

```

Which of the renewable energy generation methods have the highest capacity


```python

# Define the colors for each power source
colors = ['orange', 'red', 'green', 'Blue']

# Reshape the data into a long format
df_long = pd.melt(df, id_vars=['year'], value_vars=['Wind', 'PV', 'CSP', 'Other RE'], 
                  var_name='Power Sources', value_name='total')
fig = plt.figure(figsize=(30,10))

# Calculate the total capacity for each power source
df_pie = df_long.groupby(['Power Sources'])['total'].sum().reset_index()

# Create a pie chart with colors
plt.pie(df_pie['total'], labels=df_pie['Power Sources'], autopct='%1.1f%%', colors=colors)

# Add a title
plt.title('Total Capacity by Power Source')

# Show the plot
plt.show()

```
