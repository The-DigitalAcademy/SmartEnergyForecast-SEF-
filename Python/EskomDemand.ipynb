BUILDING THE MODEL WITH LASSO

```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error

# Load the data into a pandas dataframe
df = pd.read_csv('/Users/da_learners_m1_18/Downloads/myupdateddata.csv')

# Convert the 'datetime' column to a datetime object
df['Date Time Hour Beginning'] = pd.to_datetime(df['Date Time Hour Beginning'])

# Set the 'datetime' column as the index
df.set_index('Date Time Hour Beginning', inplace=True)

# Split the data into a features DataFrame and a target variable Series
features = df.drop('Residual Demand', axis=1)
target = df['Residual Demand']

# Split the data into a training set and a test set
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Train a Lasso regression model with alpha = 0.1
model = Lasso(alpha=0.1)
model.fit(X_train, y_train)

# Evaluate the model using the test set
score = model.score(X_test, y_test)
print("Model score using all features:", score)

# Determine the optimal number of features using k-fold cross-validation
n_features = len(X_train.columns)
cv_scores = []
for k in range(1, n_features+1):
    model = Lasso(alpha=0.1)
    scores = cross_val_score(model, X_train.iloc[:, :k], y_train, cv=5)
    cv_scores.append(scores.mean())
    # Find the number of features with the highest cross-validation score
optimal_n_features = cv_scores.index(max(cv_scores)) + 1
print("Optimal number of features:", optimal_n_features)

# Train a Lasso regression model using the optimal number of features
model = Lasso(alpha=0.1)
model.fit(X_train.iloc[:, :optimal_n_features], y_train)

# Evaluate the model using the test set
score = model.score(X_test.iloc[:, :optimal_n_features], y_test)
print("Model score using the optimal number of features:", score)

# Calculate the mean squared error (MSE) and root mean squared error (RMSE) of the model on the test set
y_pred = model.predict(X_test.iloc[:, :optimal_n_features])
mse = mean_squared_error(y_test, y_pred)
rmse = mse ** 0.5
print("MSE:", mse)
print("RMSE:", rmse)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error

# Load the data into a pandas dataframe
df = pd.read_csv("/Users/da_learner_m1_19/Downloads/SmartEnergyForecast-SEF--5/Python/myupdateddata.csv")

# Convert the 'datetime' column to a datetime object
df['Date Time Hour Beginning'] = pd.to_datetime(df['Date Time Hour Beginning'])

# Set the 'datetime' column as the index
df.set_index('Date Time Hour Beginning', inplace=True)

# Split the data into a features DataFrame and a target variable Series
features = df.drop('Residual Demand', axis=1)
target = df['Residual Demand']

# Split the data into a training set and a test set
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Train a ridge regression model with alpha = 0.1
model = Ridge(alpha=0.1)
model.fit(X_train, y_train)

# Evaluate the model using the test set
score = model.score(X_test, y_test)
print("Model score using all features:", score)

# Determine the optimal number of features using k-fold cross-validation
n_features = len(X_train.columns)
cv_scores = []
for k in range(1, n_features+1):
    model = Ridge(alpha=0.1)
    scores = cross_val_score(model, X_train.iloc[:, :k], y_train, cv=5)
    cv_scores.append(scores.mean())

# Find the number of features with the highest cross-validation score
optimal_n_features = cv_scores.index(max(cv_scores)) + 1
print("Optimal number of features:", optimal_n_features)

# Train a ridge regression model using the optimal number of features
model = Ridge(alpha=0.1)
model.fit(X_train.iloc[:, :optimal_n_features], y_train)

# Evaluate the model using the test set
score = model.score(X_test.iloc[:, :optimal_n_features], y_test)
print("Model score using the optimal number of features:", score)

# Calculate the mean squared error (MSE) and root mean squared error (RMSE) of the model on the test set
y_pred = model.predict(X_test.iloc[:, :optimal_n_features])
mse = mean_squared_error(y_test, y_pred)
rmse = mse ** 0.5
print("MSE:", mse)
print("RMSE:", rmse)
```
